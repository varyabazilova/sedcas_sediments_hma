```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import functions




def bin_elevation500(row):
    ''' function to create elevation bins''' 
    # if row['elevation'] <2500:
        # return '<2500'
    if row['elevation'] < 3000:
        return '< 3000'
    elif 3000 <= row['elevation'] < 3500:
        return '3000 - 3500'
    elif 3500 <= row['elevation'] < 4000:
        return '3500 - 4000'
    elif 4000 <= row['elevation'] < 4500:
        return '4000 - 4500'
    elif 4500 <= row['elevation'] < 5000:
        return '4500 - 5000'
    elif 5000 <= row['elevation']:
        return '> 5000'
    # elif 5500 <= row['elevation'] < 6000:
        # return '5500 - 6000'
    # elif 6000 <= row['elevation'] :
        # return '>6000'
```

```python
langtang = pd.read_csv(f'/Users/varyabazilova/Desktop/paper2/downscaling_simple/coordinates_and_elevation_with_labels_langtang.csv')[['cellnr2','band_data']].set_index('cellnr2')
mustang = pd.read_csv(f'/Users/varyabazilova/Desktop/paper2/downscaling_simple/coordinates_and_elevation_with_labels_mustang.csv')[['cellnr2','band_data']].set_index('cellnr2')

langtang = langtang.rename(columns = {'band_data':'elevation'})
mustang = mustang.rename(columns = {'band_data':'elevation'})

langtang = langtang.sort_values('elevation')
mustang = mustang.sort_values('elevation')

langtang['elevation_bin'] = langtang.apply(bin_elevation500, axis=1)
mustang['elevation_bin'] = mustang.apply(bin_elevation500, axis=1)

```
